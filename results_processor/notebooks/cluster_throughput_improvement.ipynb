```python
import pandas as pd
import itertools

# initialize list elements
data = [1,3,5,7,11, 15, 17]
read_write_ratio = [0.1, 0.25, 0.75, 0.9]

graphs = {}
for ratio in read_write_ratio:
    df = pd.DataFrame(data, columns=['cluster_size'])
    df['improvement'] = (df['cluster_size'] - 1) * ratio
    graphs[str(ratio)] = df
# Create the pandas DataFrame with column name is provided explicitly
```

```python
graphs
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
for ratio, df in graphs.items():
    # df[['cluster_size', 'improvement']].plot(ax=ax)
    # df.plot(x='cluster_size',
    #         y='improvement',
    #         ax=ax,
    #         xticks=data,
    #         legend=True,
    #         label='Read ratio - '+str(ratio),
    #         title='Read throughput improvement',
    #         ylabel='Read throughput improvement X times',
    #         xlabel='Cluster size')
    df.plot(x='cluster_size',
            y='improvement',
            ax=ax,
            xticks=data,
            legend=True,
            label='Read ratio - '+str(ratio),
            title='Read throughput improvement',
            ylabel='Read throughput improvement X times',
            xlabel='Cluster size')
    plt.plot(df['cluster_size'], df['improvement'], c='r') #connect highlighted points with a red line
    plt.scatter(df['cluster_size'], df['improvement'], marker='*', c='b') # plot specific points in red




plt.show()
```
