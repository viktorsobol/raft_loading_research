```python
import pandas as pd
def create_graph(results_data: pd.DataFrame, read_write_ratio: float):
    import matplotlib.pyplot as plt
    fig, ax = plt.subplots()

    results_data.plot(x='cluster_size',
            y=['reads_from_leader_only', 'reads_from_any_node', 'read_from_any_node_predicted'],
            ax=ax,
            xticks=results_data['cluster_size'],
            legend=True,
            label=['Always read from leader', 'Read from any node', 'Predicted result by reading from any node'],
            title='Observed read throughput improvement for ratio - {ratio}'.format(ratio=read_write_ratio),
            ylabel='RPM',
            xlabel='Cluster size')
    # plt.plot(df['cluster_size'], df['improvement'], c='r') #connect highlighted points with a red line
    plt.scatter(results_data['cluster_size'], results_data['reads_from_leader_only'], marker='*', c='r') # plot specific points in red
    plt.scatter(results_data['cluster_size'], results_data['reads_from_any_node'], marker='*', c='r') # plot specific points in red
    plt.scatter(results_data['cluster_size'], results_data['read_from_any_node_predicted'], marker='*', c='r') # plot specific points in red
    #

    plt.savefig("{ratio}_Observed_Graph_Read_throughput_improvement.eps".format(ratio=read_write_ratio), dpi=300)

    plt.show()
```
